```python
#Searching missing values
import pandas as pd

df=pd.read_csv("../data/netflix_titles.csv")
missing_values = pd.DataFrame({
    "missing_value_number":df.isna().sum(),
    "duplicated_value_number":df.duplicated().sum(),
    "missing_value_percentage":(df.isna().sum() * 100) / len(df),
    "missing_duplicated_percentage":(df.duplicated().sum() * 100) / len(df)
})
missing_values.head(12)

df=df.dropna(subset=['rating','date_added'])
df['country'] = df['country'].fillna(value=df['country'].mode()[0])

missing_values = pd.DataFrame({
    "missing_value_number":df.isna().sum(),
    "missing_value_percentage":(df.isna().sum() * 100) / len(df)
})
missing_values.head(12)


replacings=["Season","Seasons","min","s"]

for replacing in replacings:
    df['duration']=df['duration'].apply(lambda x: x.replace(replacing,""))

df["duration"]=df["duration"].astype(int)


df['date_added']=df['date_added'].apply(lambda x: x.replace(" ","-"))
df['date_added']=df['date_added'].apply(lambda x: x.replace(",",""))

df['date_added'] = pd.to_datetime(df['date_added'], format="mixed")


df.to_csv("../data/netflix_titles_cleaned.csv",index=False)



#% of missing value
#director:30,679337357
#cast:9,220495698
#country:6,510851419
#date_added:0,12841916
#rating:0,089893412

#% of duplicated value: All 0%


```
